##Exercise part 4:

```python
import pandas as pd
meteorite = pd.read_csv("Meteorite_Landings.csv")
meteorite.head()
```

```python
meteorite['year'] = meteorite['year'].str.slice(6,10)
meteorite.head()
```

```python
meteorite['year'] = meteorite['year'].apply(pd.to_numeric)
meteorite.dtypes
```

```python
meteorite=meteorite[(meteorite['year'] >= 2005) & (meteorite['year'] <= 2009)]
year = meteorite['year'].unique()
print(year)
```

```python
meteorite.set_index('year').head()
```

```python
column1 = meteorite.groupby(['year', 'fall'])['mass (g)'].quantile(0.95)
column1
```

```python
column2 = meteorite.groupby(['year', 'fall'])['mass (g)'].count()
column2
```

```python
#1
column1 = column1.to_frame()

column1["counts"] = column2
column1
```

```python
#2
meteorite.groupby("fall")["mass (g)"].describe()
```

##Exercise Part 5:

```python
import pandas as pd
taxi = pd.read_csv("2019_Yellow_Taxi_Trip_Data.csv")
taxi.head()
```

```python
taxi["tpep_dropoff_datetime"] = pd.to_datetime(taxi["tpep_dropoff_datetime"])
```

```python
taxi.set_index("tpep_dropoff_datetime", inplace=True)
```

```python
taxi.index = pd.to_datetime(taxi.index)
```

```python
taxi['hour'] = taxi.index.hour
copytaxi = taxi.groupby('hour')[['trip_distance', 'fare_amount', 'tolls_amount', 'tip_amount']].sum()
```

```python
fivetaxi = copytaxi["tip_amount"].nlargest(5)
fivetaxi
```
