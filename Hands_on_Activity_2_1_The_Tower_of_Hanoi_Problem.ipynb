```python
class TowerOfHanoi:
    #initialize the source, auxiliary and target rod
    def __init__(self):
        self.source = [ 3, 2, 1]
        self.auxiliary = []
        self.target = []
    #function that moves a disk to the target rod
    def place(self, source, target):
        disk = source.pop() #removes the last disk in the source rod
        target.append(disk) #add the disk to the target rod
        print("S =", self.source, ", A =", self.auxiliary, ", D =", self.target) #display the disk on which they have been placed

    def solve(self, n, source, target, auxiliary):
        #Only one disk to move(stopping point) it is the condition to stop the recursion if the n reaches to 1
        if n == 1:
            self.place(source, target)
            return
        # Since the If condition didnt work, this is the Step 1 which Move n-1 disks from source to auxiliary so that the largest disk is free.
        self.solve(n-1, source, auxiliary, target)
        # Step 2 which move the nth (largest) disk from source to target.
        self.place(source, target)
        # after step2 this is for to call the function again which Step 3 are to Move the n-1 disks from auxiliary to target.
        self.solve(n-1, auxiliary, target, source)


# Create an instance of the TowerOfHanoi class
torture = TowerOfHanoi()
# Solve the Tower of Hanoi problem
torture.solve(3, torture.source, torture.target, torture.auxiliary)
```

After taking one step closer to the solution, a recursive algorithm calls itself to proceed. When we arrive at the solution, the algorithm terminates.
This procedure could go on indefinitely because the called function might call itself again. In order to stop this recursion process, a base case must be provided.
